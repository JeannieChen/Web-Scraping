```python
import csv
import requests
import json
import argparse
import traceback
import numpy as np, pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def locate_stores():

    url = "https://stores.jcrew.com/en/api/v2/stores.json"

    headers = { 'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                'accept-encoding':'gzip, deflate, br',
                'accept-language':'en-US,en;q=0.5',
                'cache-control':'max-age=1800, public',
                'upgrade-insecure-requests':'1',
                'user-agent':'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:68.0) Gecko/20100101 Firefox/68.0'
    }
    stores = []
    print("retrieving stores")
    for retry in range(10):          
            stores_data = requests.get(url, headers=headers, verify=False).json().get("stores",[])
            
            if not stores_data:
                print("no stores found")
                data = {
                        "address":"",
                        "city":"",
                        "state":"",
                        "country":"",
                        "zip_code":"",
                        "store_id":"",
                        "phone":"",
                        "latitude":"",
                        "longitude":"",
                        "email":""
                }
                stores.append(data)
            
            #iterating through all stores
            for store in stores_data:
                store_id = store.get('number')
                address = store.get("url_safe_address")
                city = store.get("city")
                state = store.get("state")
                country = store.get("country_code")    
                postal_code = store.get("postal_code")           
                phone = store.get("phone_number")
                latitude = store.get("latitude")
                longitude = store.get("longitude")
                email = store.get("email")

                data = {
                        "address":address,
                        "city":city,
                        "state":state,
                        "country":country,
                        "zip_code":postal_code,
                        "store_id":store_id,
                        "phone":phone,
                        "latitude":latitude,
                        "longitude":longitude,
                        "email": email
                }
                stores.append(data)
            return stores
        #except:
            #print(trackback.format_exc())
    
    return []   
```

```python
rlt = locate_stores()
pd.DataFrame(rlt)
```

```python
df = pd.DataFrame(rlt)
df = df.loc[df.astype(str).drop_duplicates().index]
df = df.replace('', np.NaN).dropna(how='all')
df
```

```python
#export data 
df.to_csv('US_JCrew.csv', index=True, header=True)
```
