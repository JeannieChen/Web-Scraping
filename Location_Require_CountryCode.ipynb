```python
import csv
import requests
import json
import argparse
import traceback
import numpy as np, pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#https://hm.storelocator.hm.com/rest/storelocator/stores/1.0/locale/en_US/country/GB?_type=json
def locate_stores(countryCode):

    url = "https://hm.storelocator.hm.com/rest/storelocator/stores/1.0/locale/en_US/country/%s?_type=json"%(countryCode)
    
    headers = { 'accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
                'accept-encoding':'gzip, deflate, br',
                'accept-language':'en-US,en;q=0.5',
                'cache-control':'max-age=3600',
                'upgrade-insecure-requests':'1',
                'user-agent':'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:68.0) Gecko/20100101 Firefox/68.0'
    }
    stores = []
    print("retrieving stores")
    for retry in range(10):          
            stores_data = requests.get(url, headers=headers, verify=False).json().get("storesCompleteResponse",[]).get("storesComplete",[]).get("storeComplete",[])  
            
            if not stores_data:
                print("no stores found near %s"%(countryCode))
                data = {
                        "address":"",
                        "city":"",
                        "country":"",
                        "countryCode":"",
                        "store_id":"",
                        "phone":"",
                        "latitude":"",
                        "longitude":""
                }
                stores.append(data)
            
            #iterating through all stores
            for store in stores_data:
                store_id = store.get('storeId')
                address = store.get("address").get("addressLine")
                city = store.get("city")
                country = store.get("country")
                countryCode = store.get("countryCode")              
                phone = store.get("phone")
                latitude = store.get("latitude")
                longitude = store.get("longitude")

                data = {
                        "address":address,
                        "city":city,
                        "country":country,
                        "countryCode":countryCode,
                        "store_id":store_id,
                        "phone":phone,
                        "latitude":latitude,
                        "longitude":longitude
                }
                stores.append(data)
            return stores
    
    return []   
```

```python
rlt = locate_stores(countryCode='IE')
pd.DataFrame(rlt)
```

```python
df = pd.DataFrame(rlt)
df = df.loc[df.astype(str).drop_duplicates().index]
df = df.replace('', np.NaN).dropna(how='all')
df
```

```python
#export data 
df.to_csv('IE_H&M.csv', index=False, header=True)
```
